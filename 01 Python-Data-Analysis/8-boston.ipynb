```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=5)
pca.fit(boston['data'])
```

```python
Xp = pca.transform(boston['data'])
Xp.shape
```

```python
clf = RandomForestRegressor()
Xp_train, Xp_test, yp_train, yp_test = train_test_split(Xp, boston['target'], test_size=0.3)

clf.fit(Xp_train, yp_train)
clf.score(Xp_test, yp_test)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline([
    ('scale', StandardScaler()),
    ('pca', PCA(n_components=5)),
    ('svr', SVR()),
])
```

```python
X_train, X_test, y_train, y_test = train_test_split(boston['data'], boston['target'], test_size=0.3)
pipe.fit(X_train, y_train)
pipe.score(X_test, y_test)
```

```python
pipe.steps
```

```python
pipe.get_params()
```

```python
pipe.set_params(svr__C=0.9)
```

```python
import pickle
```

```python
with open('model.pickle', 'wb') as out:
    pickle.dump(pipe, out)
```

```python
with open('model.pickle', 'rb') as fp:
    pipe1 = pickle.load(fp)
```

```python
pipe1.steps
```

```python
pipe1.score(X_test, y_test)
```

```python
df = pd.DataFrame(boston['data'], columns=boston['feature_names'])
df['price'] = boston['target']

hdf = df.head(10)
hdf
```

```python
hdf.style.highlight_max()
```

```python
hdf.style.background_gradient()
```

```python
hdf.style.bar()
```

```python
def odd_green(val):
    if val % 2 != 0:
        return 'color:green'
    return ''
```

```python
hdf.style.applymap(odd_green)
```

```python
def orange10(val):
    if val > 10:
        return 'background-color:orange'
    return ''
```

```python
hdf.style.applymap(odd_green).applymap(orange10)
```

```python
hdf.style.applymap(odd_green, subset=['ZN', 'RAD']).applymap(orange10, subset=['price'])
```

```python
def top20(col):
    is_top20 = col > col.quantile(0.8)
    return ['font-weight: bold' if v else '' for v in is_top20 ]
```

```python
hdf.style.apply(top20).applymap(orange10, subset=['price'])
```
