# Basic scatter plots with matplotlib

Scatter plots can be created from Pandas Series

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
filepath = './data/Iris_Data.csv'

data = pd.read_csv(filepath)
data.head()
```

```python
plt.plot(data.sepal_length, data.sepal_width, ls='', marker='o');
```

Multiple layers of data can be added

```python
plt.plot(data.sepal_length, data.sepal_width, 
         ls='', marker='o', label='sepal')
plt.plot(data.petal_length, data.petal_width, 
         ls='', marker='o', label='petal')

plt.legend();
```

# Histograms with matplotlib

```python
plt.hist(data.sepal_length, bins=25);
```

# Customizing matplotlib plots

```python
fig, ax = plt.subplots()

ax.barh(np.arange(10), data.sepal_width.iloc[:10])

# Set position of ticks and tick labels
ax.set_yticks(np.arange(0.4, 10.4, 1.0))
ax.set_yticklabels(np.arange(1, 11))
ax.set(xlabel='Sepal width', ylabel='index', title='First 10 Iris');
```

# Incorporating statistical calculations

```python
(data.groupby('species')
    .mean()
    .plot(color=['red', 'black', 'blue', 'green'], fontsize=10.0, 
          figsize=(4,4)));
```

# Statistical plotting with seaborn

```python
import seaborn as sns
```

```python
sns.jointplot(x='sepal_length', y='sepal_width',
             data=data, size=5);
```

```python
sns.pairplot(data, hue='species', size=3);
```
