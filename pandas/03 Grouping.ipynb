# 1. Alcohol Consumption

```python
import pandas as pd


drinks = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv')
drinks.head()
```

```python
# which continent drinks more beer on average
drinks.groupby('continent').beer_servings.mean().sort_values(ascending=False)
```

```python
# statistics for mine consumption for each continent
drinks.groupby('continent').wine_servings.describe()
```

```python
# mean alcohol consumption for each column per continent
drinks.groupby('continent').mean()
```

```python
drinks.groupby('continent').median()
```

```python
drinks.groupby('continent').spirit_servings.agg(['mean', 'min', 'max'])
```

# 2. Occupation

```python
users = pd.read_table('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', 
                      sep='|', index_col='user_id')
users.head()
```

```python
# mean age per occupation
users.groupby('occupation').age.mean().sort_values(ascending=False)
```

```python
# male ratio per occupation
def gender_to_numeric(x):
    if x == 'M':
        return 1
    if x == 'F':
        return 0

users['gender_m'] = users.gender.apply(gender_to_numeric)
a = 100* (users.groupby('occupation').gender_m.sum() /
      users.occupation.value_counts())
a.sort_values(ascending=False)
```

```python
users.groupby('occupation').age.agg(['min', 'max'])
```

```python
users.groupby(['occupation', 'gender']).age.mean()
```

```python
# for each occupation present the percentage of women and men
gender_ocup = users.groupby(['occupation', 'gender']).agg({'gender': 'count'})

# create a DataFrame and apply count for each occupation
occup_count = users.groupby(['occupation']).agg('count')

# divide gender_ocup per the occup_count
occup_gender = gender_ocup.div(occup_count, level = 'occupation') * 100

occup_gender
```
