# 1. Student Alcohol consumption

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('../datasets/student-mat.csv', sep=';')
df.head()
```

```python
stud_alcoh = df.loc[:, 'school':'guardian']
stud_alcoh.head()
```

```python
# capitalize strings
capitalizer = lambda x: x.upper()

stud_alcoh.Mjob = stud_alcoh.Mjob.apply(capitalizer)
stud_alcoh.Fjob = stud_alcoh.Fjob.apply(capitalizer)
stud_alcoh.tail()
```

```python
def majority(x):
    if x > 17: 
        return True
    else:
        return False

stud_alcoh['legal_drinker'] = stud_alcoh.age.apply(majority)
stud_alcoh.head()
```

```python
stud_alcoh.dtypes
```

```python
import numpy

# multiple every number in the dataset by 10

def times10(x):
    if type(x) is str:
        return x
    elif type(x) is numpy.int64:
        return 10 * x
    else:
        return

stud_alcoh.applymap(times10).head()
```

# 2. US crime rates 1960 - 2014

```python
import numpy as np
import pandas as pd

url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/US_Crime_Rates/US_Crime_Rates_1960_2014.csv"
crime = pd.read_csv(url)
crime.head()
```

```python
crime.info()
```

```python
# convert Year to datetime64
crime.Year = pd.to_datetime(crime.Year, format='%Y')
crime.info()
```

```python
crime = crime.set_index('Year', drop=True)
crime.head()
```

```python
del crime['Total']
crime.head()
```

```python
# group the year by decades and sum the values

# resample to sum each decade
crimes = crime.resample('10AS').sum()

# get max value only for Population
population = crime['Population'].resample('10AS').max()

crimes['Population'] = population

crimes
```

```python
# most dangerous decade to live in the US
crimes.idxmax(0)
```
