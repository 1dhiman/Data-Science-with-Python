```python
def parsefile(filename):
    return np.genfromtxt(filename, 
                        delimiter= dly_delimiter,
                        usecols= dly_usecols,
                        dtype= dly_dtype,
                        names= dly_names)
```

```python
dly_delimiter = [11, 4, 2, 4] + [5, 1, 1, 1] * 31
dly_usecols = [1, 2, 3] + [4*i for i in range(1, 32)]
dly_dtype = [np.int32, np.int32, (np.str, 4)] + [np.int32] * 31
dly_names = ['year', 'month', 'obs'] + [str(day) for day in range(1, 31+1)]
```

```python
lihue = parsefile('data/USW00022536.dly')
lihue[:3]
```

```python
def unroll(record):
    startdate = np.datetime64('{}-{:02}'.format(record['year'], record['month']))
    dates = np.arange(startdate, startdate + np.timedelta64(1, 'M'), np.timedelta64(1, 'D'))
    
    rows = [(date, record[str(i+1)]/10) for i, date in enumerate(dates)]
    
    return np.array(rows, dtype=[('date', 'M8[D]'), ('value', 'd')])
```

```python
unroll(lihue[0])
```

```python
def getobs(filename, obs):
    return np.concatenate([unroll(row) for row in parsefile(filename) if row[2] == obs])
```

```python
def getobs(filename, obs):
    data = np.concatenate([unroll(row) for row in parsefile(filename) if row[2] == obs])
    
    data['value'][data['value'] == -999.9] = np.nan
    
    return data
```

```python
def fillnans(data):
    dates_float = data['date'].astype(np.float64)
    nan = np.isnan(data['value'])
    
    data['value'][nan] = np.interp(dates_float[nan], dates_float[~nan], data['value'][~nan])
```

```python
fillnans(lihue_tmax)
fillnans(lihue_tmin)
```

```python
def plot_smoothed(t, win=10):
    smoothed = np.correlate(t['value'], np.ones(win)/win, 'same')
    
    plt.plot(t['date'], smoothed)
```

```python
plot_smoothed(lihue_tmin)
```

```python
def selectyear(data, year):
    start = np.datetime64('{}'.format(year))
    end = start + np.timedelta64(1, 'Y')
    
    return data[(data['date'] >= start) & (data['date'] <= end)]['value']
```

* Find the year with the highest mean TMAX in Minneapolis
* Find the year with the lowest mean TMIN in San Diego
* In the same plot, show TMIN and TMAX throughout those year for those cities

```python
minneapolis_tmax = getobs('data/USW00014922.dly', 'TMAX')
minneapolis_tmin = getobs('data/USW00014922.dly', 'TMIN')
```

```python
sandiego_tmax = getobs('data/USW00023188.dly', 'TMAX')
sandiego_tmin = getobs('data/USW00023188.dly', 'TMIN')
```

```python
fillnans(minneapolis_tmax)
fillnans(minneapolis_tmin)

fillnans(sandiego_tmax)
fillnans(sandiego_tmin)
```

```python
years = np.arange(1940, 2014+1)

minneapolis_tmax_all = np.vstack([selectyear(minneapolis_tmax, year)[:365] for year in years])
```

```python
minneapolis_tmax_mean = np.mean(minneapolis_tmax_all, axis=1)

plt.plot(minneapolis_tmax_mean);
```

```python
minneapolis_warmest = years[np.argmax(minneapolis_tmax_mean)]
minneapolis_warmest
```

```python
sandiego_tmin_all = np.vstack([selectyear(sandiego_tmin, year)[:365] for year in years])

sandiego_tmin_mean = np.mean(sandiego_tmin_all, axis=1)

plt.plot(sandiego_tmin_mean);
```

```python
sandiego_coldest = years[np.argmin(sandiego_tmin_mean)]
sandiego_coldest
```

```python
plt.figure(figsize=(12,4))

days = np.arange(1, 367+1)

plt.fill_between(days, 
                 selectyear(minneapolis_tmin, minneapolis_warmest),
                 selectyear(minneapolis_tmax, minneapolis_warmest),
                alpha = 0.4, color = 'b')
#plt.plot(selectyear(lihue_tmin, 2009))

plt.fill_between(days, 
                 selectyear(sandiego_tmin, sandiego_coldest),
                 selectyear(sandiego_tmax, sandiego_coldest),
                alpha = 0.4, color = 'y')

plt.axis(xmin=0, xmax=367)
plt.title('{} Minneapolis vs {} San Diego'.format(minneapolis_warmest, sandiego_coldest))
plt.legend(['Minneapolis', 'San Diego']);
```
