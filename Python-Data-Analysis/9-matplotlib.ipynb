```python
import numpy as np
import pandas as pd
```

```python
xs = np.linspace(-6, 6, 100)
ys = np.sinc(xs)
```

```python
df = pd.DataFrame({'x': xs, 'sinc': ys})
```

```python
df.plot.line(x='x', y='sinc');
```

```python
import matplotlib.pyplot as plt
```

```python
plt.style.available
```

```python
plt.style.use('fivethirtyeight')
df.plot.line(x='x', y='sinc');
```

```python
plt.style.use('grayscale')
df.plot.line(x='x', y='sinc');
```

```python
import seaborn as sns
sns.set()

plt.style.use('seaborn-whitegrid')
```

```python
plt.plot(xs, ys, label=r'$\frac{sinc(x)}{x}$')
plt.legend();
```

```python
plt.plot(xs, ys, label=r'$\frac{sinc(x)}{x}$')
plt.plot(xs, np.sin(xs), label=r'$sin(x)$')

plt.legend();
```

```python
plt.subplot(2, 1, 1)
plt.plot(xs, np.sin(xs), label=r'$sin(x)$')
plt.legend()

plt.subplot(2, 1, 2)
plt.fill(xs, -xs**2, label=r'${-x}^2$')
plt.legend();

```
