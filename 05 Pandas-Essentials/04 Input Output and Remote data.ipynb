# Input and Output

```python
import pandas as pd
import numpy as np
```

### read from an Excel file

```python
file_name_string = 'data/EmployeesWithGrades.xlsx'
employees_df = pd.read_excel(file_name_string, 'Sheet1', index_col=None, na_values=['NA'])
```

```python
employees_df
```

### write to a comma separated value (.csv) file

```python
file_name_string_csv = 'data/EmployeesWithGrades.csv'
employees_df.to_csv(file_name_string_csv )
```
